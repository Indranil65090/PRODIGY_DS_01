```python
sns.barplot(x='Med. Age',y='Fert. Rate',data=df.head(10))
plt.title('Fertility Rate by Median Age of top 10 Countries')
plt.show()
```

```python
sns.histplot(df['Med. Age'],kde=True)
plt.title('Histogram for Median Age')
plt.show()
```

```python
sns.countplot(x='Med. Age',data=df.head(20))
plt.title('Count for Top 20 nation Median Age')
plt.show()
```

```python
sns.boxplot(x='Country (or dependency)',y='Population (2023)',data=df.head(6))
plt.title('Boxplot for Top 6 populous countries')
plt.show()
```

```python
sns.jointplot(x='Med. Age',y='Population (2023)',data=df,kind='hex',color='Green')
```

```python
sns.violinplot(x='Med. Age',y='Population (2023)',data=df.head(6),split=True)
plt.title('Boxplot for Top 6 populous countries')
plt.show()
```

```python
temp=df.corr()

sns.heatmap(temp,annot=True,color='Purple')
plt.title('Correlation Matrix')
plt.show()
```
